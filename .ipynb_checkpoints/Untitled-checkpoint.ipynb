```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('report_2018-2019.csv')
```

```python
print(df['Score'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['Score'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

```python
# Select every nth country for display
n = 5
subset_countries = df['Country or region'][::n].tolist()

# Visualize the data using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Score'])
plt.xlabel('Country')
plt.ylabel('Score')
plt.title('Comparison of Happiness Scores by Country')
plt.xticks(df.index[::n], subset_countries, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Extract the relevant columns
life_expectancy = df['Healthy life expectancy']
happiness_scores = df['Score']

# Visualize the data using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(happiness_scores, life_expectancy)
plt.xlabel('Happiness Score')
plt.ylabel('Life Expectancy')
plt.title('Happiness Score vs. Life Expectancy')
plt.show()
```

```python
correlation = life_expectancy.corr(happiness_scores)

print("Correlation coefficient:", correlation)
```

## We can see that there is a strong correlation between life expectancy and the happiness scores
